```python
# prompt: give me which brands are in high in visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/dairy_dataset.csv')

# Assuming 'Brand' column exists and you want to visualize its frequency
# Replace 'Brand' with the actual column name if it is different

# Calculate brand frequencies
brand_counts = data['Brand'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=brand_counts.index, y=brand_counts.values)
plt.xlabel("Brand")
plt.ylabel("Frequency")
plt.title("Brand Frequency Distribution")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()

# Alternatively, you could use a pie chart for a different visualization:
plt.figure(figsize=(8, 8))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Brand Distribution")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# prompt: give me the frequency of which brands are prefer according to location

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/dairy_dataset.csv')

# Group data by location and brand, then count occurrences
brand_location_counts = data.groupby(['Location', 'Brand']).size().unstack(fill_value=0)

# Plot the frequency of brands by location
brand_location_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Brand Preferences by Location')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Brand')
plt.tight_layout()
plt.show()
```

```python
# prompt: brands between correlation

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/dairy_dataset.csv')

# Select only numerical columns for correlation analysis
numerical_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Check if 'Brand' is in the correlation matrix columns
if 'Brand' in correlation_matrix.columns:
    # Filter for correlations related to 'Brand' (assuming 'Brand' is numerical)
    # If 'Brand' is not numerical, you'll need to adjust this part
    brand_correlations = correlation_matrix['Brand'].drop('Brand')  # Remove self-correlation

    # Print or visualize the correlations
    print("Correlations with Brand:")
    print(brand_correlations)
else:
    print("Brand column is not found in the correlation matrix. It might not be numerical or was dropped during data cleaning.")

# Create a heatmap for better visualization
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# prompt: Number of cows according to location

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/dairy_dataset.csv')

# Group data by location and sum the number of cows (assuming a 'Number of Cows' column exists)
location_cow_counts = data.groupby('Location')['Number of Cows'].sum()

# Create a bar plot
plt.figure(figsize=(10, 6))
location_cow_counts.plot(kind='bar')
plt.title('Number of Cows per Location')
plt.xlabel('Location')
plt.ylabel('Number of Cows')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Predict sales channel according to location in visualization part

# Load the dataset
data = pd.read_csv('/content/dairy_dataset.csv')

# Assuming 'Sales Channel' is the column you want to predict
# and 'Location' is the feature you want to use for prediction

# Group data by location and sales channel, then count occurrences
sales_by_location = data.groupby(['Location', 'Sales Channel']).size().unstack(fill_value=0)

# Plot the sales channel distribution for each location
sales_by_location.plot(kind='bar', stacked=True, figsize=(12, 6))  # Stacked bar chart for better comparison
plt.title('Sales Channel Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sales Channel')
plt.tight_layout()
plt.show()
```

```python
# prompt: Farm size according to locations

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/dairy_dataset.csv')

# Print the columns to see the actual column names
print(data.columns)

# Assuming 'Farm Size' column exists and contains string values like 'Small', 'Medium', 'Large'
# We need to convert these to numerical values for calculating the mean
# Create a mapping dictionary to assign numerical values to farm sizes
farm_size_mapping = {'Small': 1, 'Medium': 2, 'Large': 3}  # Adjust values as needed

# Replace farm size strings with numerical values
data['Farm Size'] = data['Farm Size'].map(farm_size_mapping)

# Group by location and calculate average farm size
farm_size_by_location = data.groupby('Location')['Farm Size'].mean()

# Create a bar plot
plt.figure(figsize=(10, 6))
farm_size_by_location.plot(kind='bar')
plt.title('Average Farm Size by Location')
plt.xlabel('Location')
plt.ylabel('Average Farm Size')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/mnt/data/dairy_dataset.csv"
df = pd.read_csv("/content/dairy_dataset.csv")

# Grouping data by Location and summing up stock levels
stock_levels_by_location = df.groupby('Location')['Quantity in Stock (liters/kg)'].sum().reset_index()

# Sorting locations by stock levels
stock_levels_by_location = stock_levels_by_location.sort_values(by='Quantity in Stock (liters/kg)', ascending=False)

# Plotting the data
plt.figure(figsize=(12, 6))
plt.bar(stock_levels_by_location['Location'], stock_levels_by_location['Quantity in Stock (liters/kg)'], color='RED')
plt.xlabel('Location')
plt.ylabel('Total Quantity in Stock (liters/kg)')
plt.title('Stock Levels by Location')
plt.xticks(rotation=90)
plt.show()

# Display the top stock levels by location
print(stock_levels_by_location.head(10))
```
