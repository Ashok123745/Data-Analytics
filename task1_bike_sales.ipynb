```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("dataset.csv")  # Update the path if needed

# Convert timestamp to datetime (if necessary)
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Set up the plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Season distribution
sns.countplot(x="season", data=df, ax=axes[0, 0], palette="Set2")
axes[0, 0].set_title("Distribution of Seasons")

# Holiday distribution
sns.countplot(x="is_holiday", data=df, ax=axes[0, 1], palette="Set2")
axes[0, 1].set_title("Distribution of Holidays")
axes[0, 1].set_xticklabels(["Non-Holiday", "Holiday"])

# Weekend distribution
sns.countplot(x="is_weekend", data=df, ax=axes[1, 0], palette="Set2")
axes[1, 0].set_title("Distribution of Weekends")
axes[1, 0].set_xticklabels(["Weekday", "Weekend"])

# Weather Code distribution
sns.countplot(x="weather_code", data=df, ax=axes[1, 1], palette="Set2")
axes[1, 1].set_title("Distribution of Weather Codes")

plt.tight_layout()
plt.show()
```

```python
# prompt: MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC.

# Assuming 'data' is your DataFrame and 'Timestamp' is the column name
# Replace 'Timestamp' with the actual name of your timestamp column if different

if 'Timestamp' in data.columns:
    data['Timestamp'] = pd.to_datetime(data['Timestamp'])
    data['DayOfWeek'] = data['Timestamp'].dt.dayofweek
    data['DayOfMonth'] = data['Timestamp'].dt.day
    data['Hour'] = data['Timestamp'].dt.hour
    data['Month'] = data['Timestamp'].dt.month
    data['Year'] = data['Timestamp'].dt.year

    # Define seasons based on month (adjust as needed for your specific location)
    def get_season(month):
        if 3 <= month <= 5:
            return 'Spring'
        elif 6 <= month <= 8:
            return 'Summer'
        elif 9 <= month <= 11:
            return 'Autumn'
        else:
            return 'Winter'

    data['Season'] = data['Month'].apply(get_season)

    data = data.set_index('Timestamp')
else:
    print("Timestamp column not found in the DataFrame.")

data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset (replace with your actual data)
data = {
    'timestamp': pd.date_range(start='2024-01-01', periods=100, freq='D'),  # 100 days starting from Jan 1, 2024
    'bike_shares': [100 + (x*10) for x in range(100)]  # Example bike shares data (increasing over time)
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract day of the week (0 = Monday, 6 = Sunday)
df['day_of_week'] = df['timestamp'].dt.day_name()

# Group by day of the week and calculate the total bike shares for each day
df_grouped = df.groupby('day_of_week')['bike_shares'].sum().reset_index()

# Sort days of the week for proper order (Monday to Sunday)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_grouped['day_of_week'] = pd.Categorical(df_grouped['day_of_week'], categories=day_order, ordered=True)

# Plotting
plt.figure(figsize=(10, 6))

# Bar plot showing bike shares by day of the week
sns.barplot(x='day_of_week', y='bike_shares', data=df_grouped)

# Title and labels
plt.title('Total Bike Shares by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bike Shares')

# Rotate x labels to make them readable
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset (replace with your actual data)
data = {
    'timestamp': pd.date_range(start='2020-01-01', periods=600, freq='D'),  # 600 days starting from Jan 1, 2020
    'bike_shares': [100 + (x*5) for x in range(600)],  # Example bike shares data (increasing over time)
    'holiday': [1 if x % 15 == 0 else 0 for x in range(600)],  # Simulate holidays every 15th day
    'season': ['Winter', 'Spring', 'Summer', 'Fall'] * 150  # Simulated seasons for simplicity
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract year from the timestamp
df['year'] = df['timestamp'].dt.year

# 1. Plotting Bike Shares by Year
plt.figure(figsize=(10, 6))

# Group by year and calculate the total bike shares for each year
df_yearly = df.groupby('year')['bike_shares'].sum().reset_index()

# Plotting the data
sns.barplot(x='year', y='bike_shares', data=df_yearly, palette='viridis')
plt.title('Total Bike Shares by Year')
plt.xlabel('Year')
plt.ylabel('Total Bike Shares')

# Show plot
plt.tight_layout()
plt.show()

# 2. Plotting Bike Shares on Holidays by Season
plt.figure(figsize=(10, 6))

# Filter data for holidays
df_holidays = df[df['holiday'] == 1]

# Group by season and calculate total bike shares on holidays
df_season_holiday = df_holidays.groupby('season')['bike_shares'].sum().reset_index()

# Plotting the data
sns.barplot(x='season', y='bike_shares', data=df_season_holiday, palette='Set2')
plt.title('Bike Shares on Holidays by Season')
plt.xlabel('Season')
plt.ylabel('Total Bike Shares on Holidays')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset (replace with your actual data)
data = {
    'timestamp': pd.date_range(start='2024-01-01', periods=100, freq='D'),  # 100 days starting from Jan 1, 2024
    'bike_shares': [100 + (x*10) for x in range(100)]  # Example bike shares data (increasing over time)
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract day of the week (0 = Monday, 6 = Sunday)
df['day_of_week'] = df['timestamp'].dt.dayofweek

# Categorize as Weekday (0-4) or Weekend (5-6)
df['week_type'] = df['day_of_week'].apply(lambda x: 'Weekday' if x < 5 else 'Weekend')

# 1. Visualizing with a Pie Chart
plt.figure(figsize=(8, 8))
bike_share_by_week_type = df.groupby('week_type')['bike_shares'].sum()
plt.pie(bike_share_by_week_type, labels=bike_share_by_week_type.index, autopct='%1.1f%%', colors=['skyblue', 'orange'])
plt.title('Distribution of Bike Shares: Weekday vs Weekend')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()

# 2. Visualizing with a Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='week_type', y='bike_shares', data=df, estimator='sum', palette='Set2')
plt.title('Total Bike Shares: Weekday vs Weekend')
plt.xlabel('Week Type')
plt.ylabel('Total Bike Shares')
plt.show()
```
