```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# Load the data from the CSV file
df = pd.read_csv('countries-continents-capitals.csv',encoding= 'latin-1')

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1("Country, Continent, and Capital Selector with Visualization"),

    # Dropdown for Countries
    html.Label("Select a Country:"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in df['Country/Territory'].unique()],
        placeholder="Select a country"
    ),

    # Dropdown for Continents
    html.Label("Select a Continent:"),
    dcc.Dropdown(
        id='continent-dropdown',
        options=[{'label': continent, 'value': continent} for continent in df['Continent'].unique()],
        placeholder="Select a continent"
    ),

    # Dropdown for Capitals
    html.Label("Select a Capital:"),
    dcc.Dropdown(
        id='capital-dropdown',
        options=[{'label': capital, 'value': capital} for capital in df['Capital'].unique()],
        placeholder="Select a capital"
    ),

    # Visualization (Bar Chart or Table)
    dcc.Graph(id='visualization'),

    # Display selected values
    html.Div(id='selected-values')
])

# Callback to update the visualization and selected values
@app.callback(
    [Output('visualization', 'figure'),
     Output('selected-values', 'children')],
    [Input('country-dropdown', 'value'),
     Input('continent-dropdown', 'value'),
     Input('capital-dropdown', 'value')]
)
def update_visualization(selected_country, selected_continent, selected_capital):
    # Filter the data based on selections
    filtered_df = df.copy()
    if selected_country:
        filtered_df = filtered_df[filtered_df['Country/Territory'] == selected_country]
    if selected_continent:
        filtered_df = filtered_df[filtered_df['Continent'] == selected_continent]
    if selected_capital:
        filtered_df = filtered_df[filtered_df['Capital'] == selected_capital]

    # Create a bar chart for visualization
    if not filtered_df.empty:
        fig = px.bar(filtered_df, x='Country/Territory', y='Capital', title="Selected Data")
    else:
        fig = px.bar(title="No Data Selected")

    # Display selected values
    selected_values = f"Selected Country: {selected_country}, Selected Continent: {selected_continent}, Selected Capital: {selected_capital}"

    return fig, selected_values

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
